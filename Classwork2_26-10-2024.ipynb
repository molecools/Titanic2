```python
df = pd.read_csv("C:\\Users\\hp\\Desktop\\python scripts\\kc_house_data.csv")

# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Calculating basic price analysis  
average_price = df['price'].mean()  
minimum_price = df['price'].min()  
maximum_price = df['price'].max()  

# Display the results  
print(f"Average Price: {average_price}")  
print(f"Minimum Price: {minimum_price}")  
print(f"Maximum Price: {maximum_price}")
```

```python


# Calculate the percentage of properties with waterfront views  
total_properties = len(df)  
waterfront_properties = df[df['waterfront'] == 1]  
waterfront_percentage = (len(waterfront_properties) / total_properties) * 100  

print(f"Percentage of properties with waterfront views: {waterfront_percentage:.2f}%")  

# Calculate average prices for waterfront and non-waterfront properties  
average_price_waterfront = waterfront_properties['price'].mean()  
average_price_non_waterfront = df[df['waterfront'] == 0]['price'].mean()  

print(f"Average price of waterfront properties: ${average_price_waterfront:,.2f}")  
print(f"Average price of non-waterfront properties: ${average_price_non_waterfront:,.2f}")  

# Plotting comparison of prices  
price_data = {  
    'Waterfront': ['Yes', 'No'],  
    'Average Price': [average_price_waterfront, average_price_non_waterfront]  
}  
price_df = pd.DataFrame(price_data)  

plt.figure(figsize=(8, 5))  
sns.barplot(data=price_df, x='Waterfront', y='Average Price', palette='viridis')  
plt.title('Average Price Comparison: Waterfront vs Non-Waterfront')  
plt.ylabel('Average Price')  
plt.yscale('log')  # Optional: Use a log scale for better price visualization  
plt.grid()  
plt.show()
```

```python
# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Renaming 'yr_renovated' to 'renovation_year' for clarity  
df['renovation_year'] = df['yr_renovated'].replace({0: None})  # Replace 0 with None for better analysis  

# Calculate the correlation between renovation year and price  
correlation = df['renovation_year'].corr(df['price'])  
print(f"Correlation between year of renovation and price: {correlation:.2f}")  

# Define a threshold for recent renovations (e.g., renovated in the last 10 years)  
recent_threshold = df['renovation_year'].max() - 10  
recently_renovated = df[df['renovation_year'] >= recent_threshold]  
not_renovated = df[df['renovation_year'].isnull()]  

# Calculate average prices  
average_price_recently_renovated = recently_renovated['price'].mean()  
average_price_not_renovated = not_renovated['price'].mean()  

print(f"Average price of recently renovated houses: ${average_price_recently_renovated:,.2f}")  
print(f"Average price of non-renovated houses: ${average_price_not_renovated:,.2f}")  

# Plotting the average prices  
price_data = {  
    'Renovation Status': ['Recently Renovated', 'Not Renovated'],  
    'Average Price': [average_price_recently_renovated, average_price_not_renovated]  
}  
price_df = pd.DataFrame(price_data)  

plt.figure(figsize=(8, 5))  
sns.barplot(data=price_df, x='Renovation Status', y='Average Price', palette='coolwarm')  
plt.title('Average Price Comparison: Recently Renovated vs Not Renovated')  
plt.ylabel('Average Price')  
plt.yscale('log')  # Optional: Use a log scale for better price visualization  
plt.grid()  
plt.show()
```

```python
# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Calculate average prices by zip code  
average_price_by_zip = df.groupby('zipcode')['price'].mean().reset_index()  

# Print the average prices for each zip code  
print("Average Prices by Zip Code:")  
print(average_price_by_zip)  

# Identify the zip codes with the highest and lowest average prices  
highest_price_zip = average_price_by_zip.loc[average_price_by_zip['price'].idxmax()]  
lowest_price_zip = average_price_by_zip.loc[average_price_by_zip['price'].idxmin()]  

print(f"Zip code with the highest average price: {highest_price_zip['zipcode']} - ${highest_price_zip['price']:,.2f}")  
print(f"Zip code with the lowest average price: {lowest_price_zip['zipcode']} - ${lowest_price_zip['price']:,.2f}")  

# Optionally, plot the average prices by zip code  
plt.figure(figsize=(14, 7))  
sns.barplot(data=average_price_by_zip.sort_values('price', ascending=False), x='zipcode', y='price', palette='viridis')  
plt.title('Average House Prices by Zip Code')  
plt.xlabel('Zip Code')  
plt.xticks(rotation=90)  # Rotate x labels for better readability  
plt.ylabel('Average Price')  
plt.yscale('log')  # Optional: Use a log scale for better price visualization  
plt.grid()  
plt.show()
```

```python
# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Calculate the age of the house  
current_year = 2023  # Update this year if necessary  
df['age_of_house'] = current_year - df['yr_built']  

# Calculate average prices by age categories (e.g., before 1950, 1950-1999, and 2000+)  
bins = [0, 25, 50, 100]  # Define age bins, adjust as needed  
labels = ['0-25 Years', '26-50 Years', '51-100 Years']  
df['age_category'] = pd.cut(df['age_of_house'], bins=bins, labels=labels, right=False)  

average_price_by_age_category = df.groupby('age_category')['price'].mean().reset_index()  

# Print average prices for age categories  
print("Average Prices by Age Category:")  
print(average_price_by_age_category)  

# Optionally, check if newer houses are priced higher than older houses  
newer_houses = df[df['age_of_house'] <= 25]  
older_houses = df[df['age_of_house'] > 25]  

average_price_newer = newer_houses['price'].mean()  
average_price_older = older_houses['price'].mean()  

print(f"Average price of newer houses (0-25 years): ${average_price_newer:,.2f}")  
print(f"Average price of older houses (26+ years): ${average_price_older:,.2f}")  

# Scatter plot to visualize the relationship between year built and price  
plt.figure(figsize=(14, 7))  
sns.scatterplot(data=df, x='yr_built', y='price', alpha=0.5)  
plt.title('Price vs Year Built')  
plt.xlabel('Year Built')  
plt.ylabel('Price')  
plt.yscale('log')  # Optional: Use a log scale for better visualization  
plt.grid()  
plt.show()
```

```python
# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Create a new column to determine if a house has a basement  
df['has_basement'] = df['sqft_basement'].apply(lambda x: 'Yes' if x > 0 else 'No')  

# Calculate average prices by basement presence  
average_price_by_basement = df.groupby('has_basement')['price'].mean().reset_index()  

# Print average prices for houses with and without basements  
print("Average Prices by Basement Presence:")  
print(average_price_by_basement)  

# Calculate price statistics (mean and standard deviation) for further insight  
mean_price_basement = average_price_by_basement.loc[average_price_by_basement['has_basement'] == 'Yes', 'price'].values[0]  
mean_price_no_basement = average_price_by_basement.loc[average_price_by_basement['has_basement'] == 'No', 'price'].values[0]  

std_dev_basement = df[df['has_basement'] == 'Yes']['price'].std()  
std_dev_no_basement = df[df['has_basement'] == 'No']['price'].std()  

print(f"Mean price of houses with basements: ${mean_price_basement:,.2f} (Std Dev: ${std_dev_basement:,.2f})")  
print(f"Mean price of houses without basements: ${mean_price_no_basement:,.2f} (Std Dev: ${std_dev_no_basement:,.2f})")  

# Box plot to visualize the price distribution by basement presence  
plt.figure(figsize=(10, 6))  
sns.boxplot(data=df, x='has_basement', y='price')  
plt.title('Price Distribution Based on Basement Presence')  
plt.xlabel('Has Basement')  
plt.ylabel('Price')  
plt.yscale('log')  # Optional: Use a log scale for better visualization of price distribution  
plt.grid()  
plt.show()
```

```python
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  

# Assuming your DataFrame is named 'df'  
# Data loading step would typically be here, e.g., df = pd.read_csv('your_file.csv')  

# Scatter plot to visualize house prices across geographical coordinates  
plt.figure(figsize=(14, 7))  
sns.scatterplot(data=df, x='long', y='lat', hue='price', size='price', sizes=(20, 200), alpha=0.5, palette='viridis')  
plt.title('House Prices Across Geographical Locations')  
plt.xlabel('Longitude')  
plt.ylabel('Latitude')  
plt.colorbar(label='Price')  
plt.grid()  
plt.show()  

# Summary statistics to check price variation by geographic coordinates  
price_stats = df[['lat', 'long', 'price']].describe()  
print(price_stats)  

# Optionally, you can group the data by locations (you can define your own method to categorize locations)  
# For practical use, you might want to define regions or neighborhoods beforehand.  
# Example: Group by rounded latitude and longitude  
df['location_category'] = df[['lat', 'long']].apply(lambda row: (round(row['lat'], 2), round(row['long'], 2)), axis=1)  

# Calculate average price per location category  
average_price_location = df.groupby('location_category')['price'].mean().reset_index()  

# Print average prices per geographical location category  
print("Average Prices by Location Category:")  
print(average_price_location)
```
