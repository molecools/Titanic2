```python
def count_keywords(text, keywords):
    if isinstance(text, str):  # Check if the text is a string
        count = 0
        for keyword in keywords:
            count += text.count(keyword)
        return count
    else:
        return 0  # Handle non-string values

# Load the dataset
df = pd.read_csv("C:\\Users\\hp\\Desktop\\python scripts\\Titanic2\\metadata.csv")

# Define the keywords to search for
keywords = ['SARS-CoV-2', 'COVID-19', 'nCoV']

# Apply the function to the 'abstract' column, handling potential errors
df['covid_mentions'] = df['abstract'].apply(lambda x: count_keywords(x, keywords))

# Print the total number of COVID-19 mentions
total_covid_mentions = df['covid_mentions'].sum()
print("Total COVID-19 mentions:", total_covid_mentions)
```
